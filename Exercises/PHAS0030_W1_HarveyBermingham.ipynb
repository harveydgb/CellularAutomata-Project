In this notebook we will explore methods of simulating and plotting the electrostatic potential and the electric field at a point due to one or more charges.

We will be using atomic units for all of the calculations. 

Following from this, the potential equation in use here will be:

$$
\phi = \frac{q}{r}
$$

Where q is the charge  and r is the scalar distance between the point and the charge.

The equation for the Electric field at the arbitrary point (x,y) is:

$$
\mathbf{E} = \frac{q}{r^2} \hat{\mathbf{r}}_{qp}
$$

Where $\hat{\mathbf{r}}_{qp}$ is the unit vector pointing in the direction from the charge to the arbitrary point (x,y).

To find the total potential or the total field at (x,y) created by more than one charge it is possible to linearly add the scalar potentials or the vector fields of each due to each of the seprate charges together for the point (x,y).

The charge arrangements explored in this document will be intitally a single charge on its own. Then we will explore the electrostatic potential and electric field created adding another charge with the opposite sign, this arrangement is called a dipole. After this we will go further and explore the field created by a quadrolpole which is a set of 4 charges arranged in a square with two of the charges having the opposite sign to the other two. The like charges are placed diagonally opposite eachother in the square.



```python
import numpy as np
import matplotlib.pyplot as plt
```

### Electrostatic Potential
#### Writing the function

```python
def elec_pot(x,y,x0,y0,q0):
    """Finds the electrostatic potential over all values given in the 2d arrays-
    of x & y due to a charge q0 at (x0,y0). 
    q0 is rounded to the nearest integer value as function uses a.u.
    All values in x & y arrays are in a.u.
    The x & y arrays must have the same shape, use np.meshgrid to make 2d.
    
    
    Intputs:
    x: 2d array giving the range of x axis values to be calculated (in a.u.)
    y: 2d array giving the range of y axis values to be calculated (in a.u.)
    x0: x coordinate of the charge in a.u.
    y0: y coordinate of the charge in a.u.
    q0: value of the charge in a.u.
    
    Outputs:
    Array of electrostatic potential values due to q0 in a.u. over all coordinates values-
    in the x&y arrays.
    """
    
    if x[i,j] in x == x0 and [i,j] in y == y0:
        print('hi')
    
    round(q0)

    #calculating array of potentials
    phi = q0/np.sqrt((x-x0)**2 + (y-y0)**2)
    
    return phi
```

#### Plotting a monopole

Here we are going to use the elec_pot() function to plot the potential over an array of x & y values for a monopole.

```python
#creating x&y arrays for all following plots to use
x = np.linspace(0,10,11)
y = np.linspace(0,10,11)
x2d, y2d = np.meshgrid(x,y)

#setting variables values for function
x1 = 2.5
y1 = 2.5
q1 = 1

pot1 = elec_pot(x2d,y2d,x1,y1,q1)

#plotting
plt.imshow(pot1, origin = 'lower')
plt.colorbar()
```

```python
#setting variables values for 2nd function
x2 = 4.5
y2 = 4.5
q2 = -1

pot2 = elec_pot(x2d,y2d,x2,y2,q2)

#plotting
plt.imshow(pot1+pot2, origin = 'lower')
plt.colorbar()
```

#### Quadropole
Now to plot the quadropole we only need to create two more potential arrays.

```python
#setting variables values for 3rd & 4th function
x3 = 2.5
y3 = 4.5
q3 = -1
x4 = 4.5
y4 = 2.5
q4 = -1

#resetting pot2 to create quadropole
q2 = 1
pot2 = elec_pot(x2d,y2d,x2,y2,q2)

pot3 = elec_pot(x2d,y2d,x3,y3,q3)
pot4 = elec_pot(x2d,y2d,x4,y4,q4)

#plotting
plt.imshow(pot1+pot2+pot3+pot4, origin = 'lower')
plt.colorbar()
```

### Electric Field
#### Writing the function

```python
def Efield(x,y,x0,y0,q0):
    """Finds the electrostatic field vector over all values given in the 2d-
    arrays of x & y due to a charge q0 at (x0,y0). 
    q0 is rounded to the nearest integer value as function uses a.u.
    All values in x & y arrays are in a.u.
    The x & y arrays must have the same shape, use np.meshgrid to make 2d.
    
    Inputs:
    x: 2d array giving the range of x axis values to be calculated (in a.u.)
    y: 2d array giving the range of y axis values to be calculated (in a.u.)
    x0: x coordinate of the charge in a.u.
    y0: y coordinate of the charge in a.u.
    q0: value of the charge in a.u.
    
    Outputs:
    Ex: x component of electric field vectors in a.u. due the charge q0 over all coordinate values-
    in the x&y arrays.
    Ey: y component of electric field vectors in a.u. due the charge q0 over all coordinate values-
    in the x&y arrays.
    """
    
    round(q0)
    
    #calculating component arrays of electric field vectors
    Ex = ((q0*(x-x0)) / ((x-x0)**2 + (y-y0)**2)**(3/2))
    Ey = ((q0*(y-y0)) / ((x-x0)**2 + (y-y0)**2)**(3/2))
    
    return Ex, Ey
```

#### Plotting a monopole

Here we are going to use the Efield function to plot the electric field vectors over all coordinate values in the x & y arrays given at the start for a monopole.

```python
#resetting x&y arrays for easier variation of electric field plots
x = np.linspace(0,10,11)
y = np.linspace(0,10,11)
x2d, y2d = np.meshgrid(x,y)

#setting variables values for function
x1 = 2.5
y1 = 2.5
q1 = 1

#creating component field arrays
Ex1, Ey1 = Efield(x2d,y2d,x1,y1,q1)

#plotting
plt.streamplot(x2d,y2d,Ex1,Ey1)
```

#### Dipole
Plotting value for a dipole.

```python
#setting variables values for function
x2 = 4.5
y2 = 4.5
q2 = -1

#creating component field arrays
Ex2, Ey2 = Efield(x2d,y2d,x2,y2,q2)

#plotting
plt.streamplot(x2d,y2d,Ex1+Ex2,Ey1+Ey2) 
```

#### Quadropole
Plotting the quadropole.

```python
#setting variables values for 3rd & 4th function
x3 = 2.5
y3 = 4.5
q3 = -1
x4 = 4.5
y4 = 2.5
q4 = -1

#resetting Ex2 & Ey2 to create quadropole
q2 = 1
Ex2, Ey2 = Efield(x2d,y2d,x2,y2,q2)

#creating component field arrays
Ex3, Ey3 = Efield(x2d,y2d,x3,y3,q3)
Ex4, Ey4 = Efield(x2d,y2d,x4,y4,q4)

#plotting
plt.streamplot(x2d,y2d,Ex1+Ex2+Ex3+Ex4,Ey1+Ey2+Ey3+Ey4)
```
